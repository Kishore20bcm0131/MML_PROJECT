```python
import os

xyz_path = "/Users/kishu/FLPCO2DB/data/raw/xyz/1/1.xyz"
clean_path = "/Users/kishu/FLPCO2DB/data/raw/xyz/1/1_clean.xyz"

with open(xyz_path) as f:
    lines = f.readlines()

# 1) First line: make sure it's just an integer
n_atoms = int(lines[0].strip())
lines[0] = f"{n_atoms}\n"

# 2) Second line: ensure there is a non-empty comment line
if len(lines) < 2:
    lines.insert(1, "FLP_1\n")
elif not lines[1].strip():
    lines[1] = "FLP_1\n"

# 3) Strip trailing whitespace on coordinate lines
for i in range(2, len(lines)):
    lines[i] = lines[i].strip() + "\n"

with open(clean_path, "w") as f:
    f.writelines(lines)

print("Wrote cleaned file to:", clean_path)
```
